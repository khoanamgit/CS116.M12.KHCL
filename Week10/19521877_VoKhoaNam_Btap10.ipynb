```python
import pandas as pd
import numpy as np
from pandas import DataFrame
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
data_link = '/gdrive/MyDrive/Social_Network_Ads.csv'
data = pd.read_csv(data_link)
```

```python
data.head()
```

```python
df = pd.DataFrame(data)
X = df.iloc[:, 0:-1];
y = df.iloc[:, -1:];
```

# NORMALIZATION DATA

```python
X_data_np = X.to_numpy()
Y_data_np = y.to_numpy()
std_X_np = StandardScaler().fit_transform(X_data_np)
std_X_df = pd.DataFrame(std_X_np)
std_X_df.head()
```

# HYPER PARAMATER OF SVM

```python
kernal = ['linear','poly','rbf','sigmoid']
rng_C = np.arange(1,52,10)
rng_co = np.arange(0.001,10,0.5)
```

# SPLIT DATA

```python
from sklearn.model_selection import train_test_split
X_train, x_test, y_train, y_test = train_test_split(std_X_np, Y_data_np, test_size=0.2,random_state = 2)
X_train_1, x_val, y_train_1, y_val = train_test_split(X_train, y_train, test_size=0.2,random_state = 2)
```

# TRAIN MODEL

```python
score = 0
for i in kernal:
  for j in rng_C:
      for n in rng_co:
          SVM_model = SVC(kernel = i, C = j,  coef0 = n )
          SVM_model.fit(X_train_1,y_train_1)
          acc_score = accuracy_score(y_val,SVM_model.predict(x_val))
          if acc_score > score:
             score = acc_score
             i1 = i
             j1 = j
             n1 = n
          
```

```python
print(score, i1, j1, n1)
```

```python
hyper_paramater = { 'kernel': ('linear','poly','rbf','sigmoid'),
             'C': [1,52,10],
             'coef0': [0.001, 10, 0.5],
           }
```

```python
SVM_model = SVC()
Grid_Search = GridSearchCV(SVM_model, hyper_paramater)
Grid_Search.fit(X_train_1,y_train_1)
```

```python
Grid_Search.best_params_
```

```python
SVM_model = SVC(kernel = 'rbf', C = 1,  coef0 = 0.001 )
SVM_model.fit(X_train_1,y_train_1)
acc_score = accuracy_score(y_test,SVM_model.predict(x_test))
```
