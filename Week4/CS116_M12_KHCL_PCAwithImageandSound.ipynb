# PCA with images

## Đọc ảnh gốc

```python
# Link chứa ảnh
pathfile_name1 = os.path.join('/drive/MyDrive/sontung1.jpg')
```

```python
# Load ảnh gốc xám
img = cv2.imread(pathfile_name1,0)
cv2_imshow(img)
```

## PCA với sklearn

```python
# PCA với thư viện sklearn
from sklearn.decomposition import PCA
pca = PCA(0.99)
print("Kích thước ảnh trước khi sử dụng PCA:",img.shape)
new_img1=pca.fit_transform(img,None)
# Kích thước của hình ảnh sau khi giảm chiều dữ liệu
print("Kích thước ảnh trước khi sử dụng PCA:",new_img1.shape)
```

## Ảnh kết quả

```python
# Load ảnh kết quả
new_img = pca.inverse_transform(new_img1)
cv2_imshow(new_img)
```
